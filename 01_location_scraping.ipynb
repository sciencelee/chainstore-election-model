## Hobby Lobby Scraping

```python

def get_hobby_lobby_loc(store_number):
    '''Looks up store number and returns long,lat'''
    URL = 'https://www.hobbylobby.com/stores/search/{}?'.format(str(store_number))
    page = requests.get(URL, allow_redirects=True)
    if page.status_code > 200:
        return
    
    soup = BeautifulSoup(page.content, 'html5lib')
    store_info = soup.find('div', class_=re.compile("store_map_details"))
    
    if store_info == None:
        return
    
    long = store_info.attrs.get('data-longitude')
    lat = store_info.attrs.get('data-latitude')
    return long, lat


# Uncomment below to scrape again

# hobby_lobby_locs = []

# for i in range(1000):
#     loc = get_hobby_lobby_loc(i)
#     hobby_lobby_locs.append(loc)
#     if i % 100 == 0:
#         print(i, loc)
    
'''
LOOKS LIKE THIS
<div class="store_map_details map-canvas" id="map_canvas"
			data-storebaseurl = "/stores/search/",
			data-latitude = "36.831171"
			data-longitude = "-119.908838"
			data-commonresourcepath = "//imgprd19.hobbylobby.com/_ui/responsive/common",
			data-stores = '[
				{
					"latitude": "36.831171",
					"longitude": "-119.908838",
					"name": "",
					"status": "",
					"statusMessage": "",
					"address2": "",
					"address1": "6485 North Riverside Dr. ",
					"city": "Fresno",
					"stateProvince": "",
					"zipcode": "93722",
					"country": "United States"
				}
			]'
		></div>

'''
```

```python
import time
import random

# first I will just grab the urls for every store.  Then I will go back and get the long/lat
tsc_urls = []
tsc_locs = []
store_seeds = ['https://www.tractorsupply.com/tsc/store_SaintJohn-IN-46373_1105',
               'https://www.tractorsupply.com/tsc/store_Holbrook-AZ-86025_1717',
               'https://www.tractorsupply.com/tsc/store_Chester-NY-10918_716',
               'https://www.tractorsupply.com/tsc/store_Starke-FL-32091_340', 
               'https://www.tractorsupply.com/tsc/store_Lufkin-TX-75904_109',
               'https://www.tractorsupply.com/tsc/store_PortOrchard-WA-98366_1923',
               'https://www.tractorsupply.com/tsc/store_Bangor-ME-04401_1353'
              ]
               
def get_tsc_stores(url, depth):
    '''Recursively finds stores by looking at map page and finding 3 closest stores, returns long lat list'''
    print(len(tsc_locs), end=" ")
    tsc_urls.append(url)
    time.sleep(random.random() * 5 + 1)
    
    page = requests.get(url, allow_redirects=True, headers={'User-Agent': 'Mozilla/5.0'})

    if page.status_code > 200:
        print("BAD")
    
    soup = BeautifulSoup(page.content, 'html5lib')

    links = []
    
    try:
        link1 = str(soup.find('form', id='storeDetails1').get('action')).replace('?zipCode=&state=&city=&returnStoreLocatorUrl=', '')
        links.append(link1)
    except:
        pass
    
    try: 
        link2 = str(soup.find('form', id='storeDetails2').get('action')).replace('?zipCode=&state=&city=&returnStoreLocatorUrl=', '')
        links.append(link2)
    except:
        pass
    
    try:
        link3 = str(soup.find('form', id='storeDetails3').get('action')).replace('?zipCode=&state=&city=&returnStoreLocatorUrl=', '')
        links.append(link3)
    except:
        pass
    
    try:
        # <div itemscope itemtype="http://schema.org/GeoCoordinates" class="hide">
		#					<meta itemprop="latitude" content="41.28437" />
		#					<meta itemprop="longitude" content="-86.62236" />
		#				</div>
    
        lat = soup.find('meta', itemprop='latitude')['content']
        long = soup.find('meta', itemprop='longitude')['content']
        if [long, lat] not in tsc_locs:
            tsc_locs.append([long, lat])
    except:
        pass
    
    garbage = '?zipCode=&city=&state=&returnStoreLocatorUrl=https://www.tractorsupply.com/tsc'
    links = [link.replace(garbage, '') for link in links]
    links = ['https://www.tractorsupply.com{}'.format(link) for link in links]
    
    
    for link in links:
        if link not in tsc_urls:
            get_tsc_stores(link, depth + 1)
    
    
for seed in store_seeds:
    get_tsc_stores(seed, 1)
    
tsc_locs
```
